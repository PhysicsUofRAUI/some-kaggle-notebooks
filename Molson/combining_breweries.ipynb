```python
import pandas as pd

#
# Creating the function to format the csv file I made into the csv file I want
#
def create_good_df(brewery_name, csv_file) :
    df = pd.read_csv(csv_file)

    df = df.drop(columns=['Unnamed: 0', '-'])

    brewery = [brewery_name] * len(df.index)

    df['Brewery'] = brewery
    
    return df

#
# Verifying the function works
#
AC_Golden_Brewing_Company = create_good_df('AC Golden Brewing Company', 'AC_Golden_Brewing_Company.csv')

AC_Golden_Brewing_Company
```

```python
#
# Collecting the rest of the data
#
Bergenbier_SA = create_good_df('Bergenbier SA', 'Bergenbier_SA.csv')
Borsodi_Brewery = create_good_df('Borsodi Brewery', 'Borsodi_Brewery.csv')
Cobra_Beer_Partnership_Limited = create_good_df('Cobra Beer Partnership Limited', 'Cobra_Beer_Partnership_Limited.csv')
Creemore_Springs_Brewery_Limited = create_good_df('Creemore Springs Brewery Limited', 'Creemore_Springs_Brewery_Limited.csv')
Franciscan_Well_Micro_Brewery = create_good_df('Franciscan Well Micro Brewery', 'Franciscan_Well_Micro_Brewery.csv')
Granville_Island_Brewery = create_good_df('Granville Island Brewery', 'Granville_Island_Brewery.csv')
Kamenitza_AD = create_good_df('Kamenitza AD', 'Kamenitza_AD.csv')
Trebjesa_Brewery = create_good_df('Trebjesa Brewery', 'Trebjesa_Brewery.csv')
Pivovary_Staropramen = create_good_df('Pivovary Staropramen', 'Pivovary_Staropramen.csv')
Molson_Coors_Canada = create_good_df('Molson Coors Canada', 'Molson_Coors_Canada.csv')
Thomas_Caffrey_Brewing_Co = create_good_df('Thomas Caffrey Brewing Co.', 'Thomas_Caffrey_Brewing_Co..csv')
Apatinska_Pivara_AD = create_good_df('Apatinska Pivara AD', 'Apatinska_Pivara_A.D..csv')
Zagrebaèka_Pivovara = create_good_df('Zagrebaèka Pivovara', 'Zagrebaèka_Pivovara_D.d..csv')
Coors_Brewing_Company = create_good_df('Coors Brewing Company (Molson-Coors)', 'Coors_Brewing_Company_(Molson-Coors).csv')
Hamms_Brewing_Co = create_good_df('Hamms Brewing Co.', 'Hamms_Brewing_Co..csv')

Hop_Valley = create_good_df('Hop Valley', 'Hop_Valley.csv')
Saint_Archer_Brewing_Co = create_good_df('Saint Archer Brewing Co.', 'Saint_Archer_Brewing_Co..csv')
Terrapin_Beer_Company = create_good_df('Terrapin Beer Company', 'Terrapin_Beer_Company.csv')
```

```python
#
# Now I will combine all of the breweries into one file
#
result = AC_Golden_Brewing_Company.append(Bergenbier_SA)

result = result.append(Borsodi_Brewery)
result = result.append(Cobra_Beer_Partnership_Limited)
result = result.append(Creemore_Springs_Brewery_Limited)
result = result.append(Franciscan_Well_Micro_Brewery)
result = result.append(Granville_Island_Brewery)
result = result.append(Kamenitza_AD)
result = result.append(Trebjesa_Brewery)
result = result.append(Pivovary_Staropramen)

result = result.append(Molson_Coors_Canada)
result = result.append(Thomas_Caffrey_Brewing_Co)
result = result.append(Apatinska_Pivara_AD)
result = result.append(Saint_Archer_Brewing_Co)
result = result.append(Zagrebaèka_Pivovara)
result = result.append(Coors_Brewing_Company)
result = result.append(Hamms_Brewing_Co)
result = result.append(Hop_Valley)
result = result.append(Terrapin_Beer_Company)

result
```
