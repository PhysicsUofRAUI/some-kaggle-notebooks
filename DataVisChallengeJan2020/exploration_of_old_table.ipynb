```python
import pandas as pd
```

```python
df_nosymbols = pd.read_csv('2010000801-noSymbol.csv', skiprows=5)
df_nosymbols = df_nosymbols.dropna()
df_nosymbols = df_nosymbols.drop(1)
```

```python
df_nosymbols["Time Period"] = pd.to_datetime(df_nosymbols["Unnamed: 0"])
```

```python
df_nosymbols = df_nosymbols.drop(columns=["Unnamed: 0"])
```

```python
df_nosymbols = df_nosymbols.replace({'Nunavut': '..'}, {'Nunavut': '0'})

df_nosymbols
```

```python
for (columnName, columnData) in df_nosymbols.iteritems():
    if columnName == "Time Period":
        continue
    df_nosymbols[columnName] = df_nosymbols[columnName].str.replace(',','')
    df_nosymbols[columnName] = pd.to_numeric(df_nosymbols[columnName])
    
df_nosymbols
```

```python
data_by_region = pd.DataFrame()
data_by_region["Quebec"] = df_nosymbols["Quebec"]
data_by_region["Ontario"] = df_nosymbols["Ontario"]
data_by_region["British Columbia"] = df_nosymbols["British Columbia"]
data_by_region["The Territories"] = df_nosymbols["Northwest Territories"] + df_nosymbols["Yukon"] + df_nosymbols["Nunavut"]
data_by_region["The Prairies"] = df_nosymbols["Saskatchewan"] + df_nosymbols["Alberta"] + df_nosymbols["Manitoba"]
data_by_region["The Atlantic Provinces"] = df_nosymbols["Newfoundland and Labrador"] + df_nosymbols["Prince Edward Island"] + df_nosymbols["Nova Scotia"] + df_nosymbols["New Brunswick"]
```

```python
data_by_region
```

```python
percent_by_region = pd.DataFrame()
percent_by_region["Time Period"] = df_nosymbols["Time Period"]
percent_by_region["Quebec"] = data_by_region["Quebec"] / df_nosymbols["Canada"]
percent_by_region["Ontario"] = data_by_region["Ontario"] / df_nosymbols["Canada"]
percent_by_region["British Columbia"] = data_by_region["British Columbia"] / df_nosymbols["Canada"]
percent_by_region["The Territories"] = data_by_region["The Territories"] / df_nosymbols["Canada"]
percent_by_region["The Prairies"] = data_by_region["The Prairies"] / df_nosymbols["Canada"]
percent_by_region["The Atlantic Provinces"] = data_by_region["The Atlantic Provinces"] / df_nosymbols["Canada"]
```

```python
percent_by_region
```

```python
example_plot_data = percent_by_region.drop(columns=["Time Period"])
example_plot_data.plot()
```
