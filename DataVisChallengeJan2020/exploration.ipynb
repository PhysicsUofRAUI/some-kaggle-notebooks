```python
df_nosymbols = pd.read_csv('2010000801-noSymbol.csv', skiprows=5)
df_nosymbols = df_nosymbols.dropna()
df_nosymbols = df_nosymbols.drop(1)
```

```python
df_nosymbols
```

```python
df_nosymbols["Canada"] = df_nosymbols["Canada"].str.replace(',','')
df_nosymbols["Newfoundland and Labrador"] = df_nosymbols["Newfoundland and Labrador"].str.replace(',','')
df_nosymbols["Prince Edward Island"] = df_nosymbols["Prince Edward Island"].str.replace(',','')
df_nosymbols["Nova Scotia"] = df_nosymbols["Nova Scotia"].str.replace(',','')
df_nosymbols["New Brunswick"] = df_nosymbols["New Brunswick"].str.replace(',','')
df_nosymbols["Quebec"] = df_nosymbols["Quebec"].str.replace(',','')
df_nosymbols["Ontario"] = df_nosymbols["Ontario"].str.replace(',','')
df_nosymbols["Manitoba"] = df_nosymbols["Manitoba"].str.replace(',','')
df_nosymbols["Saskatchewan"] = df_nosymbols["Saskatchewan"].str.replace(',','')
df_nosymbols["Alberta"] = df_nosymbols["Alberta"].str.replace(',','')
df_nosymbols["British Columbia"] = df_nosymbols["British Columbia"].str.replace(',','')
df_nosymbols["Yukon"] = df_nosymbols["Yukon"].str.replace(',','')
df_nosymbols["Northwest Territories"] = df_nosymbols["Northwest Territories"].str.replace(',','')
df_nosymbols["Nunavut"] = df_nosymbols["Nunavut"].str.replace(',','')
df_nosymbols
```

```python
df_nosymbols = df_nosymbols.replace({'Nunavut': '..'}, {'Nunavut': 0})

for (columnName, columnData) in df_nosymbols.iteritems():
    if columnName == "Unnamed: 0":
        continue
    df_nosymbols[columnName] = pd.to_numeric(df_nosymbols[columnName])
```

```python
df_nosymbols
```

```python
df_nosymbols.plot()
```

```python
df_nosymbols["The Territories"] = df_nosymbols["Northwest Territories"] + df_nosymbols["Yukon"] + df_nosymbols["Nunavut"]
```

```python
df_nosymbols = df_nosymbols.drop(columns=["Northwest Territories", "Yukon", "Nunavut"])
```

```python
df_nosymbols.plot()
```

```python
df_nosymbols["The Prairies"] = df_nosymbols["Saskatchewan"] + df_nosymbols["Alberta"] + df_nosymbols["Manitoba"]
df_nosymbols = df_nosymbols.drop(columns=["Saskatchewan", "Alberta", "Manitoba"])

df_nosymbols["The Atlantic Provinces"] = df_nosymbols["Newfoundland and Labrador"] + df_nosymbols["Prince Edward Island"] + df_nosymbols["Nova Scotia"] + df_nosymbols["New Brunswick"]
df_nosymbols = df_nosymbols.drop(columns=["Newfoundland and Labrador", "Prince Edward Island", "Nova Scotia", "New Brunswick"])
```
